```python
import cv2
import os

# Initialize empty lists to store images and ratings
images = []
ratings = []
fileNames = []


# Loop through the files in the directory
for rating in os.listdir("./images"):
    for filename in os.listdir("./images/"+rating):
        image_path = os.path.join("./images", rating,filename)
        image = cv2.imread(image_path)
        image = cv2.resize(image, (1050, 1610))
        # Perform preprocessing (e.g., resizing, normalization) as needed
#         preprocessed_image = preprocess_image(image)
        # Append the preprocessed image and rating to the lists
        images.append(image)
        fileNames.append(filename)
        ratings.append((int)(rating))

# Now, 'images' contains preprocessed images, and 'ratings' contains corresponding ratings.
```

```python
print(len(images))
print(len(ratings))
print(len(fileNames))
print(images[0].shape)
```

```python
import pandas as pd
df=pd.DataFrame(data={"image":images,"rating":ratings,"file_name":fileNames})
df
```

```python
df.to_csv('dataset.csv', index=False)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(images, ratings, test_size=0.2, random_state=42)
```

```python
import numpy as np
X_train=np.array(X_train)
X_test=np.array(X_test)
y_train=np.array(y_train)
y_test=np.array(y_test)
```

```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers

model = keras.Sequential([
    layers.Flatten(input_shape=(1610, 1050, 3)),
    layers.Dense(128, activation='relu'),
    layers.Dense(1)  # Output layer with one neuron for the rating prediction
])
```

```python
model.compile(optimizer='adam',loss='mean_squarred_error')
```

```python
model.fit(X_train,y_train,epochs=10,batch_size=16)
```
